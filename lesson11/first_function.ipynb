```python
import sympy as sp
```

```python
x = sp.symbols('x')
exp = sp.sin(x)**2 - sp.cos(x)**2
sol = sp.solveset(sp.sin(x)**2 - sp.cos(x)**2,x)
```

```python
sol
```

```python
diag = sp.plot(exp)
```

```python
exp_dif = sp.diff(exp)
exp_dif
```

```python
pos_exp_dif = sp.solveset(exp_dif > 0)
pos_exp_dif
```

```python
neg_exp_dif = sp.solveset(exp_dif < 0)
neg_exp_dif
```

```python
peak_exp_dif = sp.solveset(exp_dif)
peak_exp_dif
```

```python
pos_exp = sp.solveset(exp > 0)
pos_exp
```

```python
neg_exp = sp.solveset(exp < 0)
neg_exp
```
